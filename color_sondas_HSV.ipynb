```python
from skimage import io
import matplotlib.pyplot as plt
import matplotlib.colors as color
from skimage.measure import label, regionprops
import numpy as np
from PIL import Image
```

# **Imagen 2472**

```python
sondas = io.imread('sondas_2472_bin.jpg')

plt.imshow(sondas, cmap = 'gray')
plt.show()

for x in range(0, sondas.shape[0]):
  for y in range(0, sondas.shape[1]):
    if sondas[x][y] != 0:
      sondas[x][y] = 1

plt.imshow(sondas, cmap = 'gray')
plt.show()
```

```python
fish_2472 = io.imread("image2472.jpg")

hsv = color.rgb_to_hsv(fish_2472)

canal_h = hsv[:,:,0]

plt.imshow(canal_h, cmap='hsv')
```

```python
label_image = label(sondas)
regions = regionprops(label_image)
props = regionprops(label_image)
```

```python
masc_sondas_rojas = np.zeros(sondas.shape)
masc_sondas_verdes = np.zeros(sondas.shape)
for prop in props:
  coords = prop.centroid
  if canal_h[int(coords[0])][int(coords[1])] < 0.1 or canal_h[int(coords[0])][int(coords[1])] >= 0.9:
    for x in range(prop.bbox[0], prop.bbox[2]):
      for y in range(prop.bbox[1], prop.bbox[3]):
        if sondas[x][y] == 1:
          masc_sondas_rojas[x][y] = 1
  else:
    for x in range(prop.bbox[0], prop.bbox[2]):
      for y in range(prop.bbox[1], prop.bbox[3]):
        if sondas[x][y] == 1:
          masc_sondas_verdes[x][y] = 1
```

```python
plt.imshow(sondas, cmap = 'gray')
plt.show()
plt.imshow(masc_sondas_rojas, cmap = 'gray')
plt.show()
plt.imshow(masc_sondas_verdes, cmap = 'gray')
plt.show()
```

```python
for x in range(0, fish_2472.shape[0]):
  for y in range(0, fish_2472.shape[1]):
    if masc_sondas_rojas[x][y] != 0:
      fish_2472[x][y][0] = 255
    elif masc_sondas_verdes[x][y] != 0:
      fish_2472[x][y][1] = 255
plt.imshow(fish_2472)
plt.show()
```

```python
imagev = Image.fromarray(masc_sondas_verdes.astype(np.uint8))
imagev.save('masc_sondas_verdes2472.jpg')

imager = Image.fromarray(masc_sondas_rojas.astype(np.uint8))
imager.save('masc_sondas_rojas2472.jpg')

image = Image.fromarray(fish_2472.astype(np.uint8))
image.save('sondas2472.jpg')
```
